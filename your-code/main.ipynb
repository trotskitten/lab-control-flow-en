```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for i in range(len(words)):
    print(words[i].upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
words_5up = []
#loop in the word list, check the lenght of each word and if it's more than 4 add it to the new list
for i in range(len(words)):
    if len(words[i])>4:
        words_5up.append(words[i])
print(words_5up)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
#people means adults, kids, or both? let's implement the three possibilities
total_kids = 0
for item in people:
    total_kids += item["n_kids"]

print(f"there are {total_kids} kids and {len(people)} adults, for a total of 15 people")
```

**How many people have kids**?

```python
# your code here
#loop
pwk=[]
for item in people:
    if item["n_kids"] > 0:
        pwk.append(item['name'])
len(pwk)
#list comprehension
len([item['name'] for item in people if item["n_kids"] > 0])
```

**How many kids do they have in total?**

```python
# your code here
y = sum([item['n_kids'] for item in people])
y
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# your code here
#loop every person in the list, update each age with 1 and if their name ends with a update n_kids with 1
l_d = people
for item in l_d:
    item['age']+=1
    if item['name'].endswith("a"):
        item['n_kids']+=1
l_d


```
