## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for i in range(len(words)):
    print(words[i].upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
words_5up = []
#loop in the word list, check the lenght of each word and if it's more than 4 add it to the new list
for i in range(len(words)):
    if len(words[i])>4:
        words_5up.append(words[i])
print(words_5up)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares_to_10 = []
for i in range (11):
    squares_to_10.append(i**2)

squares_to_10
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
squares_to_10_odd = []
for i in range (11):
    if i % 2 != 0:
        squares_to_10_odd.append(i**2)

squares_to_10_odd
```
